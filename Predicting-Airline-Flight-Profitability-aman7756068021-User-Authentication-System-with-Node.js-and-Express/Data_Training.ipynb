```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv("Aviation_KPIs_Dataset.xlsx")
```

```python
df.head()
```

```python
df.isnull().isnull().sum()
```

```python
df.info()
```

```python
le = LabelEncoder()
```

```python
flot_to_number = ["Flight Number","Profit (USD)"	,"Delay (Minutes)","Aircraft Utilization (Hours/Day)","Turnaround Time (Minutes)","Load Factor (%)"	,"Fleet Availability (%)","Maintenance Downtime (Hours)","Fuel Efficiency (ASK)","Revenue (USD)","Operating Cost (USD)"	,"Net Profit Margin (%)","Ancillary Revenue (USD)","Debt-to-Equity Ratio","Revenue per ASK"	,"Cost per ASK"]
```

```python
for col in flot_to_number:
    df[col] = le.fit_transform(df[col])
```

```python
x = df[["Flight Number"	,"Delay (Minutes)","Aircraft Utilization (Hours/Day)","Turnaround Time (Minutes)","Load Factor (%)"	,"Fleet Availability (%)","Maintenance Downtime (Hours)","Fuel Efficiency (ASK)","Revenue (USD)","Operating Cost (USD)"	,"Net Profit Margin (%)","Ancillary Revenue (USD)","Debt-to-Equity Ratio","Revenue per ASK"	,"Cost per ASK"]]
y = df[["Profit (USD)"]]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.5, random_state=45)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
y_pred
```

```python
print("R² Score (multi-target):", r2_score(y_test, y_pred, multioutput='uniform_average'))
```
